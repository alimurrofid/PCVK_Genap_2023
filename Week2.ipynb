Membuat black image dengan tipe data int16

```python
black_img = np.zeros(shape=(512,512,3), dtype = np.int16)
plt.imshow(black_img)
```

Membuat bentuk persegi panjang di koordinat (384, 0) dan (510, 150)

```python
#perhatikan koordinat titik2 pt1 dan pt2
cv.rectangle(black_img, pt1=(384,0), pt2=(510,150), color=(0,255,0), thickness=10)
plt.imshow(black_img)
```

Menambahkan bentuk persegi pada koordinat (200, 200) dan (300, 300)

```python
cv.circle(black_img, center=(100,100), radius=50, color=(255,0,0), thickness=8)
plt.imshow(black_img)
```

2. Apakah perbedaan dan pengaruhnya pembuatan black image antara tipe data int16
dan int32?


*   Perbedaan antara tipe data int16 dan int32 adalah dalam ukuran penyimpanan data. int16 menggunakan 16 bit, sementara int32 menggunakan 32 bit. Ini mempengaruhi kapasitas rentang nilai yang dapat disimpan dalam gambar. Gambar hitam pada int16 lebih efisien dalam penggunaan memori, sementara int32 memiliki kapasitas yang lebih besar untuk nilai-nilai yang dapat disimpan, tetapi membutuhkan lebih banyak memori.

5. Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image.
disarankan menggunakan bilangan acak/random!

```python
import random
canvas = np.zeros((300,300,3),dtype=np.uint8)
boxes = 5
for _ in range(boxes):
    box_width = random.randint(30, 100)
    box_height = random.randint(30, 100)

    top_left_x = random.randint(0, 300 - box_width)
    top_left_y = random.randint(0, 300 - box_height)

    color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))

    cv.rectangle(canvas, (top_left_x, top_left_y), (top_left_x + box_width, top_left_y + box_height), color, -1)

cv2_imshow(canvas)
```
